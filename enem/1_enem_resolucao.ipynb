```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

enem = pd.read_csv('enem.csv')
```

```python
# 7. Análise exploratória
# a) Qual o percentual de pessoas que possui internet em casa (['Q025'].map({'A' : 'Não', 'B': 'Sim'}))?
# b) Qual a escolaridade da mãe dos candidatos (['Q002'].map({'A':'Nunca estudou', 'B':'Não completou EF', 'C':'Não completou EF', 'D':'Não completou EM', 'E':'Não completou ES', 'F':'Graduada', 'G':'Pós-graduada','H':'Não sabe'}))?
# c) Qual a renda familiar (['Q006'].map({'A':'Nenhuma renda', 'B':'Até 1.320', 'C':'Até 1.320', 'D':'Até 3.520', 'E':'Até 3.520', 'F':'Até 3.520', 'G':'Até 3.520', 'H':'Até 10.560','I':'Até 10.560', 'J':'Até 10.560','K':'Até 10.560', 'L':'Até 10.560','M':'Até 10.560','N':'Até 10.560', 'O':'Mais de 10.560','P':'Mais de 10.560','Q':'Mais de 10.560'}))?
# d) Qual o histograma de idades (matplotlib hist)?
# e) Como se comparam as notas de CH com MT?
# f) Qual a correlação entre as notas de CH e MT?
# g) Como se compara a renda familiar com a nota em MT?
# h) Como se compara a escolaridade da mãe com a nota em MT?
# i) Como se compara o local de residência com a nota em MT?
```

```python
#a) Qual o percentual de pessoas que possui internet em casa (['Q025'].map({'A' : 'Não', 'B': 'Sim'}))?

enem['Q025'] = enem['Q025'].map({'A' : 'Não', 'B': 'Sim'})
enem['Q025'].value_counts() / len(enem) * 100
```

```python
import seaborn as sns
sns.countplot(enem['Q025'])
#usando seaborn
```

```python
# b) Qual a escolaridade da mãe dos candidatos (['Q002'].map({'A':'Nunca estudou', 'B':'Não completou EF', 'C':'Não completou EF', 'D':'Não completou EM', 'E':'Não completou ES', 'F':'Graduada', 'G':'Pós-graduada','H':'Não sabe'}))?

enem['Q002'] = enem['Q002'].map({'A':'Nunca estudou', 'B':'Não completou EF', 'C':'Não completou EF', 'D':'Não completou EM', 'E':'Não completou ES', 'F':'Graduada', 'G':'Pós-graduada','H':'Não sabe'})
enem['Q002'].value_counts() / len(enem) * 100
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(enem['Q002'])
```

```python
# c) Qual a renda familiar (['Q006'].map({'A':'Nenhuma renda', 'B':'Até 1.320', 'C':'Até 1.320', 'D':'Até 3.520', 'E':'Até 3.520', 'F':'Até 3.520', 'G':'Até 3.520', 'H':'Até 10.560','I':'Até 10.560', 'J':'Até 10.560','K':'Até 10.560', 'L':'Até 10.560','M':'Até 10.560','N':'Até 10.560', 'O':'Mais de 10.560','P':'Mais de 10.560','Q':'Mais de 10.560'}))?

enem['Q006'] = enem['Q006'].map({'A':'Nenhuma renda', 'B':'Até 1.320', 'C':'Até 1.320',
                             'D':'Até 3.520', 'E':'Até 3.520', 'F':'Até 3.520', 'G':'Até 3.520',
                             'H':'Até 10.560','I':'Até 10.560', 'J':'Até 10.560','K':'Até 10.560',
                             'L':'Até 10.560','M':'Até 10.560','N':'Até 10.560',
                             'O':'Mais de 10.560','P':'Mais de 10.560','Q':'Mais de 10.560'})
plt.figure(figsize=(12, 6))
sns.countplot(enem['Q006'])
print(enem['Q006'].value_counts() / len(enem) * 100)
```

```python
# d) Qual o histograma de idades (matplotlib hist)?

plt.figure(figsize=(12, 6))
plt.hist(enem['NU_IDADE'], bins=7, histtype='bar')
plt.show()
```

```python
# e) Como se comparam as notas de CH com MT?

# enem_df['NU_NOTA_CH'].isna().sum()
plt.hist(enem['NU_NOTA_CH'], alpha=0.7, color='red')
plt.hist(enem['NU_NOTA_MT'], alpha=0.7, color='blue')
plt.show()
```

```python
# f) Qual a correlação entre as notas de CH e MT?

plt.figure(figsize=(12, 6))
plt.scatter(enem['NU_NOTA_CH'], enem['NU_NOTA_MT'], alpha=0.5)
plt.xlabel('Ciencias Humanas')
plt.show()
```

```python
# g) Como se compara a renda familiar com a nota em MT?

plt.figure(figsize=(12, 6))
sns.boxplot(x=enem['Q006'], y=enem['NU_NOTA_MT'])
plt.show()
```

```python
# h) Como se compara a escolaridade da mãe com a nota em MT?

plt.figure(figsize=(12, 6))
sns.boxplot(x=enem['Q002'], y=enem['NU_NOTA_MT'])
plt.show()
```

```python
# i) Como se compara o local de residência com a nota em MT?

plt.figure(figsize=(12, 6))
sns.boxplot(x=enem['SG_UF_RESIDENCIA'], y=enem['NU_NOTA_MT'])
plt.show()
```
